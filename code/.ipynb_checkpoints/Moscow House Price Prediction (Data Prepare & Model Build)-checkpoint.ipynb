#### FUNCTIONS

```python
from sklearn import metrics

def regression_report(dependent_test, 
                      estimator, 
                      independent_test,
                      predictions_plot=True):
    """ 
    This function calculates and prints basic performance metrics for regression evaluation.
    Additionally, it plots predicted vs actual values if `predictions_plot` is True.

    Parameters:
        dependent_test (array-like): Ground truth target values (y_test).
        estimator (model): Trained regression model.
        independent_test (array-like): Test input data (X_test).
        predictions_plot (bool): Whether to plot predicted vs actual values. Default is True.
    """
    
    # Calculate predictions
    predictions = estimator.predict(independent_test)
    
    # Performance Metrics
    print(f'R2 : {metrics.r2_score(dependent_test, predictions)}')
    print(f'MAE : {metrics.mean_absolute_error(dependent_test, predictions)}')
    print(f'MSE : {metrics.mean_squared_error(dependent_test, predictions)}')
    print(f'RMSE : {np.sqrt(metrics.mean_squared_error(dependent_test, predictions))}')
    
    # Plotting Predictions
    if predictions_plot:
        plt.figure(figsize=(12, 6))
        plt.scatter(dependent_test, predictions, alpha=0.6, edgecolor='k', label='Predictions')
        plt.plot([dependent_test.min(), dependent_test.max()],
                 [dependent_test.min(), dependent_test.max()],
                 color='red', linestyle='--', label='Ideal Fit')
        plt.xlabel('Actual Values')
        plt.ylabel('Predicted Values')
        plt.title('Actual vs Predicted Values')
        plt.legend()
        plt.grid(alpha=0.3)
```

## CLEAN & PREPARE

#### PREPARE DATA FOR MODEL BUILD 

```python
moscow_housing_data_independent = moscow_housing_data.select_dtypes(include = ('number')).drop(columns = ['price'])

moscow_housing_data_dependent = moscow_housing_data['price']


moscow_housing_data_independent_train , moscow_housing_data_independent_test , moscow_housing_data_dependent_train , moscow_housing_data_dependent_test = train_test_split(
    moscow_housing_data_independent ,
    moscow_housing_data_dependent ,
    train_size = 0.95 ,
    random_state = 42
)
```

```python
for array in [moscow_housing_data_independent_train,moscow_housing_data_independent_test]:

    print(f'{array.shape}')
```

## MODEL BUILD

#### HGB

```python
from sklearn.ensemble import HistGradientBoostingRegressor

moscow_housing_price_hgb = HistGradientBoostingRegressor()

moscow_housing_price_hgb_param_distributions = {
                                                'learning_rate': [0.01, 0.05, 0.1, 0.2],  # Step size
                                                'max_iter': [100, 300, 500],  # Number of boosting iterations
                                                'max_depth': [None, 3, 5, 7],  # Maximum tree depth
                                                'min_samples_leaf': [10, 20, 50],  # Minimum samples per leaf
                                                'l2_regularization': [0, 0.1, 1],  # L2 regularization
                                                'max_bins': [128, 255],  # Valid range: 2 to 255
                                                }
```

```python
from sklearn.model_selection import RandomizedSearchCV , cross_val_score

moscow_housing_price_hgb_random_search_cv = RandomizedSearchCV( estimator = moscow_housing_price_hgb ,
                                                                param_distributions = moscow_housing_price_hgb_param_distributions,
                                                                scoring = 'neg_mean_squared_error',
                                                                n_iter = 25,
                                                                cv = 10 )
```

```python
moscow_housing_price_hgb_random_search_cv.fit( moscow_housing_data_independent_train,
                                               moscow_housing_data_dependent_train )
```

```python
moscow_housing_price_hgb_random_search_cv.best_score_
```

```python
moscow_housing_price_hgb_random_search_cv.best_params_
```

```python
moscow_housing_price_hgb_random_search_cv_cross_val_scores = cross_val_score( estimator = HistGradientBoostingRegressor(l2_regularization=0.1, 
                                                                                                                        max_iter=500,
                                                                                                                        min_samples_leaf=10,
                                                                                                                        max_depth = None,
                                                                                                                        learning_rate = 0.1,
                                                                                                                        max_bins = 255) ,
                                                                              X = moscow_housing_data_independent_train,
                                                                              y = moscow_housing_data_dependent_train ,
                                                                              scoring = 'r2',
                                                                              cv=5)


```

```python
moscow_housing_price_hgb_random_search_cv_mean_cross_val_score = np.round(np.mean(moscow_housing_price_hgb_random_search_cv_cross_val_scores),2)


moscow_housing_price_hgb_random_search_cv_mean_cross_val_score
```

```python
moscow_housing_price_hgb_random_search_cv_cross_val_scores
```

#### ExtraTrees

```python
from sklearn.ensemble import ExtraTreesRegressor

moscow_housing_price_extra_trees = ExtraTreesRegressor()

moscow_housing_price_extra_trees_param_distributions = {
    'n_estimators': [100, 300, 500],  # Number of trees
    'max_depth': [None, 10, 20, 30],  # Tree depth
    'min_samples_split': [2, 5, 10],  # Min samples to split
    'min_samples_leaf': [1, 2, 5, 10],  # Min samples per leaf
    'max_features': ['sqrt', 'log2', None],  # Feature selection per split
    'bootstrap': [False, True]  # Bootstrapping
}
```

```python
moscow_housing_price_extra_trees_random_search_cv = RandomizedSearchCV( estimator = moscow_housing_price_extra_trees ,
                                                                        param_distributions = moscow_housing_price_extra_trees_param_distributions,
                                                                        scoring = 'neg_mean_squared_error',
                                                                        n_iter = 25,
                                                                        cv = 7 )
```

```python
moscow_housing_price_extra_trees_random_search_cv.fit( moscow_housing_data_independent_train,
                                                       moscow_housing_data_dependent_train )
```

```python
moscow_housing_price_extra_trees_random_search_cv.best_score_
```

```python
moscow_housing_price_extra_trees_random_search_cv.best_params_
```

```python
moscow_housing_price_extra_trees_random_search_cv_cross_val_scores = cross_val_score( estimator = ExtraTreesRegressor(n_estimators=300,
                                                                                                                      min_samples_leaf=1,
                                                                                                                      min_samples_split = 2,
                                                                                                                      max_depth = None,
                                                                                                                      max_features = 'sqrt',
                                                                                                                      bootstrap = False) ,
                                                                              X = moscow_housing_data_independent_train,
                                                                              y = moscow_housing_data_dependent_train ,
                                                                              scoring = 'r2',
                                                                              cv=5)
```

```python
moscow_housing_price_extra_trees_random_search_cv_mean_cross_val_score = np.round(np.mean(moscow_housing_price_extra_trees_random_search_cv_cross_val_scores),2)


moscow_housing_price_extra_trees_random_search_cv_mean_cross_val_score
```

```python
moscow_housing_price_extra_trees_random_search_cv_cross_val_scores
```

### TEST EVALUATION

```python
moscow_housing_price_hgb = HistGradientBoostingRegressor(
    **moscow_housing_price_hgb_random_search_cv.best_params_
)

moscow_housing_price_extra_trees = ExtraTreesRegressor(
    **moscow_housing_price_extra_trees_random_search_cv.best_params_
)
```

```python
for estimator in [moscow_housing_price_hgb,moscow_housing_price_extra_trees]:

    estimator.fit(moscow_housing_data_independent_test ,
                  moscow_housing_data_dependent_test),

    print(estimator)
    regression_report( estimator = estimator,
                       independent_test = moscow_housing_data_independent_test,
                       dependent_test = moscow_housing_data_dependent_test,
                       predictions_plot = True )
```
